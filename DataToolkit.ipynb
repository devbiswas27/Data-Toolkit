#Q.1. What is NumPy, and why is it widely used in Python?

NumPy (Numerical Python) is a fundamental open-source library for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

Why NumPy is Widely Used

Here are the key reasons for its popularity:

1. Powerful N-dimensional Array Object (ndarray)

NumPy‚Äôs ndarray is a fast, flexible container for large datasets in Python.

It enables vectorized operations, avoiding the need for explicit loops in many cases.

2. Performance

NumPy is implemented in C and optimized for performance.

It can be orders of magnitude faster than pure Python, especially for large data operations.

3. Broadcasting Capabilities

Broadcasting allows NumPy to perform operations on arrays of different shapes intelligently without the need for manual replication of data.

4. Extensive Collection of Mathematical Functions

Includes linear algebra, statistical, Fourier transform, and random number capabilities.

5. Foundation for Other Libraries

Many popular data science and machine learning libraries (like Pandas, SciPy, Scikit-learn, TensorFlow, and PyTorch) use NumPy arrays internally.

It acts as the base layer for numerical computing in Python.

6. Ease of Integration

Easily integrates with C, C++, and Fortran code, making it ideal for high-performance computing.

Example Usage:

    import numpy as np

    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])

    print(a + b)  # [5 7 9]
    print(np.dot(a, b))  # Dot product: 32

# Q. 2.How does broadcasting work in NumPy?

Broadcasting in NumPy allows operations between arrays of different shapes by automatically stretching or replicating the smaller array so that they have compatible shapes for element-wise operations.

üß† Concept: How Broadcasting Works

When NumPy performs operations on two arrays, it compares their shapes element-wise from right to left.

Two dimensions are compatible if:

They are equal, or

One of them is 1

If the shapes can be aligned, NumPy broadcasts the smaller array to match the larger one.

‚úÖ Steps NumPy Uses for Broadcasting

Given two arrays A and B:

Compare their shapes from right to left.

If dimensions match or one is 1, they are compatible.

NumPy virtually stretches the array with 1 to match the other (without copying data).

The operation is performed element-wise.

#Q.4. Explain the use of the groupby() method in Pandas?

The groupby() method in Pandas is used to split a DataFrame into groups based on one or more keys (columns), and then apply a function (like sum(), mean(), count(), etc.) to aggregate or transform those groups.

üß† Conceptual Workflow

groupby() performs:

Splitting ‚Äì Split the data into groups based on a key or multiple keys.

Applying ‚Äì Apply a function to each group independently.

Combining ‚Äì Combine the results back into a single structure.

üìä Basic Example

    import pandas as pd

    data = {
        'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'],
        'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Salary': [50000, 55000, 60000, 65000, 70000]
    }

    df = pd.DataFrame(data)

# Group by 'Department' and get average salary
    grouped = df.groupby('Department')['Salary'].mean()
    print(grouped)
    Output:
    Department
    Finance    70000.0
    HR         52500.0
    IT         62500.0
    Name: Salary, dtype: float64

#Q.8.What does the term ‚Äúvectorized operation‚Äù mean in NumPy?

The term "vectorized operation" in NumPy refers to performing operations on entire arrays (vectors, matrices, etc.) without using explicit loops in Python.

Instead of looping through elements one-by-one, vectorized operations apply a function or operation to the entire array at once, leveraging highly optimized C-level code under the hood.

‚úÖ Why Vectorized Operations Matter

Faster: Much more efficient than Python loops

Cleaner code: No need to write complex for-loops

Less error-prone: Operates on whole data structures

üìä Example: Without vs. With Vectorization

üê¢ Python Loop (Non-vectorized)

    a = [1, 2, 3]
    b = [4, 5, 6]
    result = []

    for i in range(len(a)):
        result.append(a[i] + b[i])

    print(result)  # [5, 7, 9]
‚ö° NumPy Vectorized

    import numpy as np

    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])

    print(a + b)  # [5 7 9]

#Q.15. How does NumPy handle multidimensional arrays?

NumPy handles multidimensional arrays using its core data structure called the ndarray (N-dimensional array), which allows you to store and manipulate arrays of any number of dimensions (1D, 2D, 3D, and beyond) efficiently.

 What Is a Multidimensional Array?

A multidimensional array is simply an array of arrays. For example:

1D: [1, 2, 3]

2D: [[1, 2, 3], [4, 5, 6]]

3D: A cube-like structure (array of matrices), etc.

>> Key Concepts in NumPy‚Äôs Multidimensional Array Handling

 1. Creation

You can create multidimensional arrays using np.array():


import numpy as np

# 2D array (matrix)
arr_2d = np.array([[1, 2], [3, 4]])

# 3D array (array of matrices)
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

 2. Attributes

        Attribute	Description
        ndim	Number of dimensions
        shape	Tuple showing size along each dimension
        size	Total number of elements
        dtype	Data type of elements
        print(arr_3d.ndim)   # 3
        print(arr_3d.shape)  # (2, 2, 2)

 3. Indexing and Slicing

Just like with lists, but for multiple axes:


# Access element
arr_3d[0, 1, 1]  # Element in first matrix, second row, second column

# Slice across axes
arr_3d[:, 0, :]  # First row from all matrices

 4. Vectorized Operations on Multidimensional Arrays
Operations are element-wise across all dimensions:

        arr_2d * 10      # Multiplies every element by 10
        np.sqrt(arr_3d)  # Applies square root to all elements
5. Reshaping and Transposing

arr.reshape(4, 2)      # Reshape to new shape
arr.transpose(1, 0)    # Swap axes


 6. Broadcasting

NumPy can perform operations between arrays of different shapes by broadcasting smaller arrays to match the shape of larger ones:


      a = np.array([[1], [2], [3]])  # Shape (3, 1)
      b = np.array([10, 20])         # Shape (2,)
      a + b  # Result shape: (3, 2)

#Q.17. Explain the difference between apply() and map() in Pandas?

>> Key Difference

      Feature	             map()	                     apply()

      Applies to	      Each element in a Series	     Rows/columns (in DataFrame) or elements (in Series)

A dictionary for value replacement

Another Series for mapping


    s.map({1: 'A', 2: 'B'})  # Replaces 1 with 'A', 2 with 'B'

 2. apply() ‚Äì For Series and DataFrames

More versatile.

Can be used on:

Series (like map())

DataFrames, either row-wise or column-wise

    s.apply(lambda x: x ** 2)

      df = pd.DataFrame({

      })

#Q.19. How does Pandas simplify time series analysis?

Pandas simplifies time series analysis by offering a rich set of tools specifically designed for working with date- and time-based data. It allows for efficient parsing, indexing, resampling, windowing, and visualization of time series, making complex operations straightforward.

>> Key Ways Pandas Simplifies Time Series Analysis

1.  Datetime Indexing
Pandas allows you to use DatetimeIndex as the index of a DataFrame or Series.

    import pandas as pd

    dates = pd.date_range('2023-01-01', periods=6, freq='D')

    ts = pd.Series([10, 20, 15, 30, 25, 40], index=dates)

This enables:

Date-based slicing: ts['2023-01-03':'2023-01-05']

Time-aware resampling and rolling calculations

2. Date Parsing and Conversion

Pandas can automatically parse string dates to datetime objects.

    pd.to_datetime(['2023-01-01', '2023-02-01'])
Works even with mixed formats, making messy data easy to clean.

3. Resampling

Resample time series to a different frequency (e.g., daily to monthly).


    ts.resample('M').mean()  # Monthly average
You can also use:

.sum(), .median(), .ohlc() (open-high-low-close), etc.

4.  Rolling Window Calculations

Apply operations over a rolling window (e.g., moving average).

    ts.rolling(window=3).mean()
This is essential for smoothing noisy data or computing trends.

5.  Time Offsets and Shifting

Shift data forward or backward in time.

    ts.shift(1)         # Shift values down by 1 period
    ts.tshift(freq='D') # Shift timestamps (deprecated in newer versions)
6.  Datetime Properties

Access components like year, month, day, weekday easily:

    ts.index.month
    ts.index.day_name()
7. Time Zone Handling

You can localize or convert time zones:

    ts.tz_localize('UTC').tz_convert('US/Eastern')
8.  Time Series Filtering and Slicing

Pandas allows intuitive time-based slicing:


    ts['2023']            # All of 2023
    ts['2023-01']         # January 2023
    ts['2023-01-02']      # Specific day

#Q.21.Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing?

NumPy‚Äôs array slicing is faster and more memory-efficient than Python‚Äôs list slicing due to how NumPy arrays are implemented under the hood.

Here‚Äôs a detailed breakdown of why NumPy slicing is faster:

 1. Contiguous Memory Storage

NumPy arrays are stored in a contiguous block of memory.

Python lists store references to objects scattered throughout memory.

 Result:

NumPy can access and manipulate data more efficiently using pointer arithmetic.

No overhead of following object references like in Python lists.

 2. Slicing Creates Views, Not Copies

In NumPy, slicing an array creates a view (not a copy) of the original data.

In contrast, Python list slicing creates a new list, copying all the elements.

    import numpy as np
    a = np.array([1, 2, 3, 4])
    b = a[1:3]     # View in NumPy (no data copied)

‚úÖ This means:

Less memory usage

Faster performance

Modifying the view can affect the original array (if intended)

 3. Vectorized C Implementation

NumPy is implemented in C, and many operations, including slicing, are handled at the low level.

Python list slicing is interpreted in Python and involves more overhead.

 NumPy slicing is handled in compiled C code, making it extremely fast.

 4. Fixed-Type, Homogeneous Elements

NumPy arrays are homogeneous (same data type), which allows slicing logic to operate with predictable element sizes.

Python lists are heterogeneous, requiring type checking and pointer dereferencing during slicing.

‚è± Performance Comparison

    import numpy as np
    import time

    # NumPy array
    arr_np = np.arange(10**6)
    start = time.time()
    slice_np = arr_np[100000:200000]
    print("NumPy slice time:", time.time() - start)

    # Python list
    arr_list = list(range(10**6))
    start = time.time()
    slice_list = arr_list[100000:200000]
    print("List slice time:", time.time() - start)

#Q.22.What are some common use cases for Seaborn?

Seaborn is a powerful data visualization library built on top of Matplotlib, and it's particularly well-suited for statistical data visualization. It provides high-level interfaces to create attractive, informative, and statistically rich visualizations. Below are some common use cases for Seaborn, showcasing its ability to make complex visualizations simple and clear.

>>Common Use Cases for Seaborn

1. Visualizing Distributions

Seaborn provides several functions to visualize the distribution of data, which is crucial in understanding the spread, central tendency, and outliers.

Histograms (sns.histplot)

KDE (Kernel Density Estimation) (sns.kdeplot)

Box plots (sns.boxplot)

Violin plots (sns.violinplot)

Example: Distribution Plot

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    tips = sns.load_dataset("tips")

    # Distribution of total bill
    sns.histplot(tips['total_bill'], kde=True)
    plt.show()
Use Case: Understanding the distribution of a numerical variable like sales, temperature, or income.

2. Correlation Heatmaps

Seaborn makes it easy to visualize the correlation matrix between multiple variables using heatmaps (sns.heatmap).

Example: Correlation Heatmap

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    iris = sns.load_dataset("iris")

    # Calculate correlation matrix
    corr = iris.corr()

    # Heatmap of correlation matrix
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
    plt.show()
Use Case: Detecting relationships between features (e.g., in a dataset of financial indicators, scientific measurements, etc.).

3. Categorical Data Visualization

Seaborn provides powerful functions for visualizing categorical data. You can easily compare categories in a bar plot, box plot, strip plot, or swarm plot.

Bar plots (sns.barplot)

Box plots (sns.boxplot)

Swarm plots (sns.swarmplot)

Count plots (sns.countplot)

Example: Boxplot for Categorical Comparison

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    tips = sns.load_dataset("tips")

    # Boxplot for total bill by day of the week
    sns.boxplot(x='day', y='total_bill', data=tips)
    plt.show()
Use Case: Comparing performance across different groups or categories, such as sales performance per region or test scores across different student groups.

4. Pairwise Relationships (Pair Plots)

Seaborn‚Äôs pairplot allows you to visualize pairwise relationships in a dataset, which is useful for spotting correlations or patterns across multiple features.

Example: Pairplot

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    iris = sns.load_dataset("iris")

    # Pairplot of the dataset
    sns.pairplot(iris, hue='species')
    plt.show()
Use Case: Exploring relationships between features like age, income, and spending in a dataset, or height, weight, and age in a health dataset.

5. Regression Plots

Seaborn simplifies the creation of linear regression plots that help visualize the relationship between two continuous variables, including the regression line and confidence intervals.

Linear regression plots (sns.regplot)

Polynomial regression (sns.regplot with order argument)

Example: Regression Plot

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    tips = sns.load_dataset("tips")

    # Regression plot for total bill vs tip
    sns.regplot(x='total_bill', y='tip', data=tips)
    plt.show()
Use Case: Analyzing the relationship between variables such as price and demand, temperature and sales, or hours worked and performance.

6. Facet Grids for Subsetting Data

Seaborn provides FacetGrid to create a grid of subplots, allowing you to visualize the data across multiple subsets (e.g., based on different categorical features).

Example: FacetGrid for Subsets of Data

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load a dataset
    tips = sns.load_dataset("tips")

    # Create a facet grid for day-wise distribution of total bill
    g = sns.FacetGrid(tips, col="day")
    g.map(sns.histplot, "total_bill")
    plt.show()
Use Case: Visualizing trends across different categories, such as sales by region, student performance by course, or weather patterns by city.

7. Time Series Analysis

Seaborn allows easy plotting of time series data, often used in economic, stock market, or sensor data analysis.

Example: Time Series Plot

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load dataset (a simple time series dataset)
    flights = sns.load_dataset("flights")

    # Pivot the data for time series plotting
    pivot_flights = flights.pivot("month", "year", "passengers")

    # Plot the time series data
    sns.lineplot(data=pivot_flights)
    plt.show()
Use Case: Plotting trends over time, such as stock prices, sales over months, or temperature trends.

8. Visualization of Missing Data

Seaborn also makes it easy to visualize missing data, which can be useful for data preprocessing.

Example: Visualizing Missing Data

    import seaborn as sns
    import matplotlib.pyplot as plt

    # Load dataset with missing data
    tips = sns.load_dataset("tips")

    # Introduce missing values
    tips.loc[0:5, 'total_bill'] = None

    # Heatmap for missing values
    sns.heatmap(tips.isnull(), cbar=False, cmap='viridis')
    plt.show()

```python
#Q.11. Perform matrix multiplication using NumPy
# Matrix A (2x3 matrix)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Matrix B (3x2 matrix)
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])
# Using np.dot()
result_dot = np.dot(A, B)

# Alternatively, using the @ operator
result_at = A @ B

# Display the result
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix multiplication result (using np.dot()):")
print(result_dot)

print("\nMatrix multiplication result (using @):")
print(result_at)
```
